```python
# Histogram to visualize distribution of songs by length
plt.figure(figsize=(10, 5))
sns.histplot(df["len"], bins=100, kde=True)
plt.axvline(df["len"].mean(), color="blue", linestyle="solid", label="Mean")
plt.axvline(df["len"].median(), color="red", linestyle="dashed", label="Median")
plt.legend
plt.xlabel("Length of Songs")
plt.ylabel("Frequency")
plt.title("Histogram of Song Length")
plt.show()
```

```python
hip_hop_df = df[df['genre'] == 'hip hop']
sns.histplot(hip_hop_df['release_date'], kde=True)
plt.title('Distribution of Release Years for Hip Hop Genre')
plt.xlabel('Release Year')
plt.ylabel('Frequency')
plt.show()
```

## Bivariate Analysis

```python
plt.figure(figsize=(14, 6))
sns.violinplot(data=df, x='genre', y='obscene', palette='rocket', inner='quartile')

plt.xticks(rotation=45)
plt.title('Obscene Scores by Genre')
plt.xlabel('Genre')
plt.ylabel('Obscene Score')
plt.tight_layout()
plt.show()
```

```python
# Box plot of Genre vs Release Decade
plt.figure(figsize=(14, 6))
sns.boxplot(
    data=df,
    x='release_date',
    y='genre',
    palette='Set2'
)
plt.title('Genre vs Release Decade', fontsize=16)
plt.xlabel('Release Decade', fontsize=12)
plt.ylabel('Genre', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
pivot = df.pivot_table(index='genre', columns='topic', aggfunc='size')

pivot.plot(kind='bar', stacked=False, figsize=(15,5))
plt.title("Distribution of Topics Across Genres")
plt.xlabel("Genre")
plt.ylabel("Number of Songs")
plt.legend(title="Topic", bbox_to_anchor=(1, 1))
plt.show()
```

## Multivariate Analysis

```python
df.columns
```

```python
# One-hot encode the 'genre' and 'topic' columns
df_encoded = pd.get_dummies(df, columns=['genre', 'topic'])

# Sample 1% of the data for testing purposes
df_sampled = df_encoded.sample(frac=0.01, random_state=42)

# Compute the correlation matrix for the sampled data (include both 'genre' and 'topic')
corr_matrix = df_sampled[['age', 'release_date', 'len'] + 
                         [col for col in df_encoded.columns if 'genre_' in col] + 
                         [col for col in df_encoded.columns if 'topic_' in col]].corr()

# Plot heatmap
plt.figure(figsize=(16, 10))
sns.heatmap(corr_matrix, annot=True, cmap='PuBu', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of Features (Genre and Topic Included)')
plt.show()
```

```python
# Set style
sns.set(style="whitegrid")

# Create scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='feelings', y='sadness', alpha=0.6)

# Calculate and display correlation coefficient
correlation = df['feelings'].corr(df['sadness'])
plt.title(f'Sadness vs. Feelings (Correlation: {correlation:.2f})', fontsize=14)
plt.xlabel('Feelings Score')
plt.ylabel('Sadness Score')
plt.tight_layout()
plt.show()
```

```python
# Define list of emotion-related variables
emotion_vars = [
    'romantic', 'world/life', 'family/spiritual', 'communication',
    'shake the audience', 'dating', 'violence', 'movement/places',
    'light/visual perceptions'
]

# Create scatter plots of feelings vs each emotional category
plt.figure(figsize=(14, 12)) 
for i, var in enumerate(emotion_vars):
    plt.subplot(3, 3, i + 1)
    sns.regplot(
        data=df,
        x='feelings',
        y=var,
        scatter_kws={'alpha': 0.4},
        line_kws={'color': 'red'}
    )
    plt.title(f"Feelings vs {var.replace('/', ' ').capitalize()}", fontsize=12)
    plt.xlabel("Feelings", fontsize=10)
    plt.ylabel(var.replace('/', ' ').capitalize(), fontsize=10)
    plt.xticks
plt.tight_layout()
plt.show()
```
